```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Imports
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import TimeDistributed, LSTM, Dense, GlobalAveragePooling2D, Dropout, BatchNormalization, Input
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.utils.class_weight import compute_class_weight
import matplotlib.pyplot as plt

# CONFIG
IMG_SIZE = 224
SEQUENCE_LENGTH = 5
BATCH_SIZE = 4
EPOCHS = 30
base_path = "/content/drive/MyDrive/Traffic_Violation_Dataset"
class_names = ['helmet', 'no_helmet', 'overloading']

# Load sequences
def load_sequence_from_folder(folder, seq_len=SEQUENCE_LENGTH):
    sequences, labels = [], []
    label_map = {name: idx for idx, name in enumerate(class_names)}
    for class_name in class_names:
        class_folder = os.path.join(folder, class_name)
        class_images = sorted([os.path.join(class_folder, img)
                               for img in os.listdir(class_folder) if img.endswith(('.jpg', '.jpeg', '.png'))])
        for i in range(0, len(class_images) - seq_len + 1, seq_len):
            sequences.append(class_images[i:i + seq_len])
            labels.append(label_map[class_name])
    return sequences, labels

# Data Augmentation
datagen = ImageDataGenerator(rotation_range=10, width_shift_range=0.1,
                             height_shift_range=0.1, horizontal_flip=True)

# Sequence Generator
class SequenceGenerator(tf.keras.utils.Sequence):
    def __init__(self, sequences, labels, batch_size, is_train=True):
        self.sequences = sequences
        self.labels = labels
        self.batch_size = batch_size
        self.is_train = is_train
        self.on_epoch_end()

    def __len__(self):
        return len(self.sequences) // self.batch_size

    def on_epoch_end(self):
        self.indices = np.arange(len(self.sequences))
        if self.is_train:
            np.random.shuffle(self.indices)

    def __getitem__(self, index):
        batch_indices = self.indices[index*self.batch_size:(index+1)*self.batch_size]
        batch_sequences = [self.sequences[i] for i in batch_indices]
        batch_labels = [self.labels[i] for i in batch_indices]

        X = np.zeros((self.batch_size, SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3))
        for i, seq_paths in enumerate(batch_sequences):
            for j, img_path in enumerate(seq_paths):
                img = load_img(img_path, target_size=(IMG_SIZE, IMG_SIZE))
                img = img_to_array(img) / 255.0
                if self.is_train:
                    img = datagen.random_transform(img)
                X[i, j] = img
        y = tf.keras.utils.to_categorical(batch_labels, num_classes=3)
        return X, y

# Load data
train_seq, train_labels = load_sequence_from_folder(os.path.join(base_path, "Train"))
val_seq, val_labels = load_sequence_from_folder(os.path.join(base_path, "Validation"))
test_seq, test_labels = load_sequence_from_folder(os.path.join(base_path, "Test"))

# Class weights
class_weights = compute_class_weight(class_weight='balanced', classes=np.unique(train_labels), y=train_labels)
class_weights_dict = dict(enumerate(class_weights))

# Generators
train_gen = SequenceGenerator(train_seq, train_labels, BATCH_SIZE)
val_gen = SequenceGenerator(val_seq, val_labels, BATCH_SIZE, is_train=False)
test_gen = SequenceGenerator(test_seq, test_labels, BATCH_SIZE, is_train=False)

# Base model
cnn_base = MobileNetV2(include_top=False, weights='imagenet', input_shape=(IMG_SIZE, IMG_SIZE, 3))
cnn_base.trainable = True
for layer in cnn_base.layers[:-30]:
    layer.trainable = False

cnn_model = Sequential([
    cnn_base,
    GlobalAveragePooling2D(),
    Dropout(0.4),
    BatchNormalization()
])

# CNN-LSTM model
model = Sequential([
    TimeDistributed(cnn_model, input_shape=(SEQUENCE_LENGTH, IMG_SIZE, IMG_SIZE, 3)),
    LSTM(48, return_sequences=False, kernel_regularizer=l2(0.01), dropout=0.4, recurrent_dropout=0.2),
    Dropout(0.5),
    Dense(64, activation='relu', kernel_regularizer=l2(0.01)),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

# Compile
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),
              loss='categorical_crossentropy', metrics=['accuracy'])

# Callbacks
callbacks = [
    EarlyStopping(patience=5, restore_best_weights=True),
    ModelCheckpoint("/content/drive/MyDrive/best_model.h5", save_best_only=True)
]

# Train
history = model.fit(train_gen, validation_data=val_gen, epochs=EPOCHS,
                    class_weight=class_weights_dict, callbacks=callbacks)

# Evaluate
loss, acc = model.evaluate(test_gen)
print(f"Test Accuracy: {acc:.4f}, Test Loss: {loss:.4f}")

# Plots
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title("Accuracy over Epochs")
plt.legend()
plt.show()

plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title("Loss over Epochs")
plt.legend()
plt.show()
```

```python
# Gather all test data at once (much faster than looping)
X_test, y_test = zip(*[test_gen[i] for i in range(len(test_gen))])
X_test = np.concatenate(X_test)
y_test = np.concatenate(y_test)

# Predict once for all test data
y_prob = model.predict(X_test, batch_size=BATCH_SIZE, verbose=1)
y_pred = np.argmax(y_prob, axis=1)
y_true = np.argmax(y_test, axis=1)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import seaborn as sns

# 1. Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

# 2. Classification Report
report = classification_report(y_true, y_pred, target_names=class_names)
print("Classification Report:\n", report)

# 3. ROC Curve for each class
plt.figure(figsize=(8, 6))
for i in range(len(class_names)):
    fpr, tpr, _ = roc_curve(y_test[:, i], y_prob[:, i])
    plt.plot(fpr, tpr, label=f"{class_names[i]} (AUC = {roc_auc_score(y_test[:, i], y_prob[:, i]):.2f})")
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

plt.figure(figsize=(8, 6))
for i in range(len(class_names)):
    precision, recall, _ = precision_recall_curve(y_test[:, i], y_prob[:, i])
    ap = average_precision_score(y_test[:, i], y_prob[:, i])
    plt.plot(recall, precision, label=f"{class_names[i]} (AP = {ap:.2f})")

plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.metrics import f1_score

f1_scores = f1_score(y_true, y_pred, average=None)
plt.figure(figsize=(6, 4))
sns.barplot(x=class_names, y=f1_scores)
plt.title("F1 Score per Class")
plt.ylabel("F1 Score")
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```

```python
from sklearn.metrics import precision_score, recall_score
import pandas as pd

precision = precision_score(y_true, y_pred, average=None)
recall = recall_score(y_true, y_pred, average=None)
f1 = f1_score(y_true, y_pred, average=None)
support = np.bincount(y_true)

metrics_df = pd.DataFrame({
    'Class': class_names,
    'Precision': precision,
    'Recall': recall,
    'F1-Score': f1,
    'Support': support
})

print(metrics_df.to_string(index=False))
```
